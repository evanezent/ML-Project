```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.externals import joblib 
```

```python
data = pd.read_csv('classification-datasets.csv')
```

Cek apakah ada missing value pada data

```python
missing_values = data.isnull().sum()
missing_values
```

```python
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
%matplotlib inline  
import matplotlib as mpl
```

```python
def boxPlot(data1, data2):
    mpl.rcParams['figure.dpi'] = 100
    plt.figure(figsize=(60, 60))
    f, axes = plt.subplots(1, 2)
    sns.boxplot(y=data1, ax=axes[0])
    sns.boxplot(y=data2, ax=axes[1])
    plt.subplots_adjust(wspace=1)
```

```python
boxPlot(data["height_cm"], data["weight_kg"])
```

#### Formula to find out upper limit and lower limit from Outliers

```python
def setbound(datacolumn):
    sorted(datacolumn)
    Q1,Q3 = np.percentile(datacolumn , [25,75])
    IQR = Q3 - Q1
    print(Q1, Q3)
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range,upper_range
```

#### Get the outliers data

```python
def getOutlierData(data, mainData):
    lowerbound,upperbound = setbound(data)
    return mainData[(data < lowerbound) | (data > upperbound)]
```

#### Drop the outliers data

```python
def outliering(data1, data2, mainData):
    while True:
        lowerbound,upperbound = setbound(data1)
        res1 = getOutlierData(data1, mainData)

        lowerbound,upperbound = setbound(data2)
        res2 = getOutlierData(data2, mainData)
        
        res = pd.concat([res1,res2]).drop_duplicates()
        res = res.index

        print('Total Outliers = ',len(res))
        print('=== The Outliers ===')
        print(res)    
        mainData.drop(res, inplace=True)

        if(len(res)==0):
            break
    print("BLOXPOT AFTER OUTLIERING")
    boxPlot(mainData['height_cm'], mainData['weight_kg'])
    return mainData
```

```python
newData = outliering(data["height_cm"], data["weight_kg"], data)
```

```python
newData
```

#### Mengambil Independen Data

```python
data_train = []
for item in newData.values:
#     print(item[1])
    data_train.append([item[0], item[1]])
data_train
```

```python
train = np.array(data_train)
```

#### Mengambil Dependen Data

```python
label = np.array(newData['body_type'])
```

### Split the Dataset

```python
#split dataset 80% dan 20% secara random tiap n random state data
x_train, x_test, y_train, y_test = train_test_split(train, label, test_size = 0.2, random_state = 100)
```

## Naive Bayes

```python
from sklearn.naive_bayes import ComplementNB
```

#### Make Naive Bayes Classifier

```python
classifierx = ComplementNB()
```

#### Predict datatest using the NB Model

```python
classifierx.fit(x_train, y_train)
```

```python
hasilx = classifierx.predict(x_test)
```

#### Evaluate Model

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, hasilx))
```

#### Saving Naive Bayes Model

```python
joblib.dump(classifierx, "classification-NB.h5")
```

### Kesimpulan
Dari data di atas, didapatkan bahwa menggunakan metode Naive Bayes mendapatkan akurasi mencapai 60%

## SVM

```python
from sklearn import svm
```

#### Make SVM Classifier

```python
classifier2 = svm.SVC(decision_function_shape='ovo')
classifier2.fit(x_train, y_train)
```

#### Predict datatest using the SVM Model

```python
hasil2 = classifier2.predict(x_test)
```

#### Evaluate Model

```python
print(classification_report(y_test, hasil2))
```

#### Saving SVM Model

```python
joblib.dump(classifier2, "classification-SVM.h5")
```

### Kesimpulan
Dari data di atas, didapatkan bahwa menggunakan metode SVM mendapatkan akurasi mencapai 62%
