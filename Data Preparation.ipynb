```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("fifa20.csv")
```

```python
data = pd.DataFrame(data, columns=['sofifa_id', 'player_url', 'age', 'height_cm', 'weight_kg', 'body_type', 'overall', 'potential'])
```

```python
data
```

```python
data.describe()
```

```python
import seaborn as sns
sns.set_style("whitegrid")
```

### Melihat korelasi antar data

```python
corr = data.select_dtypes(include = ['float64', 'int64']).iloc[:, :].corr()
plt.figure(figsize=(10, 10))
ax = sns.heatmap(corr, vmax=1, square=True)
# ax.set_xticklabels(rotation=30)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
```

## Pemotongan Data

<h4>Data untuk <b>Body Type</b> <i>Classification</i> berdasarkan <b>Height</b> dan <b>Weight</b></h4>

```python
csv_clas = pd.DataFrame(data, columns=['height_cm', 'weight_kg', 'body_type'])
csv_clas
```
